```python
device = "cpu"

model.to(device)

def generate_answer(model, tokenizer, prompt_template, max_tokens=100, temperature=0.5, top_k=30, top_p=0.9, num_return_sequences=1):
    input_ids = tokenizer(prompt_template, return_tensors='pt').input_ids.cpu()
    # output = model.generate(inputs=input_ids, do_sample=True, max_new_tokens=1, temperature=temperature, top_k=top_k, top_p=top_p, num_return_sequences=num_return_sequences)
    output = model.generate(inputs=input_ids, max_new_tokens=1)
    return tokenizer.decode(output[0])

def post_process_answer(generated_text):

    answer_text = generated_text.split(' assistant')[-1]

    # stop_keywords = ['therefore', 'in conclusion', 'thus', 'hence', 'consequently']

    # sentences = answer_text.split('. ')

    # last_complete_sentence = None
    # for sentence in reversed(sentences[:-1]):
    #     if any(keyword in sentence.lower() for keyword in stop_keywords):
    #         last_complete_sentence = sentence
    #         break

    # if last_complete_sentence is None:
    return answer_text

    # answer = answer_text.split(last_complete_sentence)[0] + last_complete_sentence
    # return answer
```

```python
import json

def inference(system_message, model, tokenizer, input_file, output_file, context=False):
    short_answers = {}

    with open(input_file, 'r') as f:
        data = json.load(f)

        for k, v in data.items():
            question = v['QUESTION']

            if context:
                contexts = v['CONTEXTS']
                input = " Contexts: " + " ".join(contexts) + " Question: " + question
            else:
                input = question

            prompt_template=f'''<|im_start|>system
            {system_message}<|im_end|>
            <|im_start|>user
            {input}<|im_end|>
            <|im_start|>assistant
            '''

            generated_text = generate_answer(model, tokenizer, prompt_template)
            processed_text = post_process_answer(generated_text)
            # print(f"{k}: \t{processed_text}")

            short_answers[k] = processed_text

    with open(output_file, 'w') as f:
        json.dump(short_answers, f)
```

```python
system_message = "Answer the question in either yes, no or maybe."
# With context
input_file = 'test_set.json'
output_file = 'test_results_1.json'

inference(system_message, model, tokenizer, input_file, output_file, context=True)
```
